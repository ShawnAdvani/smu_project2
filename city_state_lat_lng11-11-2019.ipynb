```python
import pandas as pd
import requests
import json

# remove this before pushing and add config folder to hold key
gkey = "AIzaSyA9oRdShW29ntgNVInE7zmqCHVBWCok63E"
```

```python
path = "citystate.csv"
cities_df = pd.read_csv(path)
cities_df.head()
```

```python
city_list = cities_df["RegionName"] + ", " + cities_df["State"]
city_list.head()
```

```python
lat = []
lng = []

for city in city_list:
   
        target_url = ('https://maps.googleapis.com/maps/api/geocode/json?address={0}&key={1}').format(city, gkey)
    
        geo_data = requests.get(target_url).json()
    
        lat.append(geo_data["results"][0]["geometry"]["location"]["lat"])
        lng.append(geo_data["results"][0]["geometry"]["location"]["lng"])
    
        print('''
        City: {0}
        Latitude: {1}
        Longitude: {2}
        '''.format(city, (geo_data["results"][0]["geometry"]["location"]["lat"]), geo_data["results"][0]["geometry"]["location"]["lng"]))

```

```python
cities_df["Latitude"] = lat
cities_df["Longitude"] = lng
cities_df.head()
```

```python
cities_df.to_csv("city_state_lat_lng.csv")
```
